## Ordinal Regression data

```python
from fri import genOrdinalRegressionData
X,y = genOrdinalRegressionData(n_samples=300, n_features=8,n_strel=1, n_redundant=2,
                    n_repeated=0,n_target_bins=3,noise=0.2,random_state=123)
```

```python
X.shape
```

We created a regression set with 6 features of which 2 are strongly relevant and 2 weakly relevant.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

Scale Data

```python
from sklearn.preprocessing import StandardScaler
X_scaled = StandardScaler().fit_transform(X)
```

New object for Regression Data

```python
from fri import FRIOrdinalRegression
fri_model = FRIOrdinalRegression()
```

Fit to data

```python
fri_model.fit(X_scaled,y)
```

Print out feature relevance intervals

```python
fri_model.interval_
```

### Plot results

```python
from fri.plot import plot_relevance_bars
fig, ax = plt.subplots(1, 1,figsize=(12,6))
out = plot_relevance_bars(ax,fri_model.interval_,classes=fri_model.relevance_classes_)
```

### Print internal Parameters

```python
fri_model.allrel_prediction_
```
